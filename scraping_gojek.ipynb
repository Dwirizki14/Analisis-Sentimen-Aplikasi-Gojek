```python
!pip install google-play-scraper
from google_play_scraper import app, reviews, Sort
import pandas as pd
import os
```

```python
# Scraping ulasan Gojek
result, _ = reviews(
    'com.gojek.app',  # Package name Gojek di Play Store
    lang='id',        # Bahasa Indonesia
    country='id',     # Lokasi Indonesia
    sort=Sort.NEWEST, # Ambil ulasan terbaru
    count=10000       # Target jumlah ulasan
)
```

```python
# Konversi ke DataFrame
reviews_df = pd.DataFrame(result)
reviews_df
```

```python
# save to csv
reviews_df.to_csv('gojek_reviews.csv', index=False)
```
